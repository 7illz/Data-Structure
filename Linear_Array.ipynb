```python
#TASK 4
def balanceSalami(salami):
    l=len(salami)
    x=int(l/2)
    sum=0
    j=0
    rev=len(salami)-1

    for i in range(x):
        sum+=salami[i]
    for i in range(x):
        j+=salami[rev]
        rev-=1
    if sum>j:
        j+=salami[x]

    else:
        sum+=salami[x]

    if sum==j:
            return True
    else:
            return False

salami=np.array([1, 1, 1, 2, 1])
print(balanceSalami(salami))
```

```python
import numpy as np
def calcu_mean(a1):
    sum=0
    for i in range(len(a1)):
        sum+=a1[i]
    return (sum/len(a1))

def calcu_sd(a1):
    mean=calcu_mean(a1)
    sum=0
    for i in range(len(a1)):
        sum+=((a1[i]-mean)**2)
    ans=(sum/(len(a1)-1))**(1/2)
    ans=round(ans,2)

    return ans

def away_from_sd(a1):
    mean=calcu_mean(a1)
    sd=calcu_sd(a1)
    away1=mean+((1.5)*sd)
    away2=mean-((1.5)*sd)
    new=0
    for i in range(len(a1)):
        if a1[i]>away1 or a1[i]<away2:
            new+=1
    arr=np.array([0]*new)
    count=0
    for i in range(len(a1)):
        if a1[i]>away1 or a1[i]<away2:
            arr[count]=a1[i]
            count+=1
    return arr
arr=np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(calcu_mean(arr))
print(calcu_sd(arr))
print(away_from_sd(arr))
```
