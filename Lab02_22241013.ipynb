```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  #To Do
  row=len(floor)
  col=len(floor[0])
  for i in range(col):
    if i%2==0:

        for j in range(0,row,2):
            print(floor[j][i], end=' ')
    else:
        if row%2!=0:
            for j in range(row-2,0,-2):
                print(floor[j][i], end=' ')
        else:
            for j in range(row-1,0,-2):
                print(floor[j][i],end=' ')
    print()




floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    #TO DO
    col=len(district[0])
    row=len(district)
    nrow=row+(depth*2)
    ncol=col+(depth*2)
    mat=np.zeros((nrow,ncol),dtype=int)
    for i in range(0,nrow,1):
        for j in range(0,ncol,1):
            mat[i][j]=8
    for i in range(depth,row+depth,1):
        for j in range(depth,col+depth,1):
            mat[i][j]=district[i-depth][j-depth]
    return mat

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

```python
def strength_difference(clubroom):
  strength_diff = np.zeros(100, dtype = int)
  #To Do
  row=len(clubroom)
  col=(len(clubroom[0]))
  size=((row*col)-row)//2
  strength_diff=[0]*size
#   return strength_diff
  x=0
  for i in range(row):
    diff=0
    for j in range(0,col,1):
        if i!=j and j>i:
            diff=clubroom[i][j]-clubroom[j][i]
            strength_diff[x]=diff
            x+=1
  return np.array(strength_diff)


clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  #pass
  mat=np.zeros((n,n),dtype=int)
  for i in range(n):
    sum=0
    for j in range((i+1)):
        if i==j or j==0:
            mat[i][j]=i+1
        elif j==1:
            mat[i][j]=mat[i-1][j]+mat[i-1][j-1]
            sum=mat[i-1][j]+mat[i-1][j-1]
        else:
            mat[i][j]=sum+mat[i-1][j]
            sum=mat[i][j]

  return mat

def print_without_none(matrix):
  #this prints the non None elements of matrix
  for i in range(len(matrix)):
    for j in range(len(matrix)):
        if matrix[i][j]!=0:
            print((matrix[i][j]),end=' ')
    print()

n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

```python
def moving_around(cmds):
    grid = np.full((7, 7), '.')
    i,j=3,3
    for x in cmds:
        if x==1:
            if 0<=(i-2)<=6 and 0<=(j-3)<=6:
                i,j=i-2,j-3
                grid[i][j]="*"
        if x==2:
            if 0<=(i-2)<=6 and 0<=(j-1)<=6:
                i,j=i-2,j-1
                grid[i][j]="*"
        if x==3:
            if 0<=(i-3)<=6 and 0<=(j-2)<=6:
                i,j=i-3,j-2
                grid[i][j]="*"
        if x==4:
            if 0<=(i-1)<=6 and 0<=(j-2)<=6:
                i,j=i-1,j-2
                grid[i][j]="*"
        if x==5:
            if 0<=(i-2)<=6 and 0<=(j+1)<=6:
                i,j=i-2,j+1
                grid[i][j]="*"
        if x==6:
            if 0<=(i-2)<=6 and 0<=(j+3)<=6:
                i,j=i-2,j+3
                grid[i][j]="*"
        if x==7:
            if 0<=(i-3)<=6 and 0<=(j+2)<=6:
                i,j=i-3,j+2
                grid[i][j]="*"
        if x==8:
            if 0<=(i-1)<=6 and 0<=(j+2)<=6:
                i,j=i-1,j+2
                grid[i][j]="*"
        if x==9:
            if 0<=(i+2)<=6 and 0<=(j-3)<=6:
                i,j=i+2,j-3
                grid[i][j]="*"
        if x==10:
            if 0<=(i+2)<=6 and 0<=(j-1)<=6:
                i,j=i+2,j-1
                grid[i][j]="*"
        if x==11:
            if 0<=(i+1)<=6 and 0<=(j-2)<=6:
                i,j=i+1,j-2
                grid[i][j]="*"
        if x==12:
            if 0<=(i+3)<=6 and 0<=(j-2)<=6:
                i,j=i+3,j-2
                grid[i][j]="*"

    grid[3][3]="-"
    grid[i][j]='/'
    return grid

cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
```
