```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def smallest_level(root,level=0,dic={}):
    if root==None:
        return

    if root!=None and level not in dic.keys():
        dic[level]=root.elem

    if level in dic.keys():
        if root.elem< dic[level]:
            dic[level]=root.elem

    smallest_level(root.left,level+1,dic)
    smallest_level(root.right,level+1,dic)
    return dic



#DRIVER CODE
root = tree_construction([None, 4,9,2,3,-5,None,7])
print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  3 9 5 4 2 7
print()
print('Level Wise Smallest Value: ', end = ' ')
print(smallest_level(root)) #Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def inorder_predecessor(root, x):

    if root==None:
        return None


    elif root.elem>=x.elem:
        return inorder_predecessor(root.left,x)

    elif root.elem<x.elem:
        right=inorder_predecessor(root.right,x)

        if right==None:
            return root
        else:
            return right


#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
```

```python
### 2
def create_BST(arr,i,end):

    if i>end:
        return

    mid=(i+end)//2

    root=BTNode(arr[mid])

    root.left=create_BST(arr,i,mid-1)
    root.right=create_BST(arr,mid+1,end)

    return root





arr=[1,2,3,4,5,6,7]
var=len(arr)-1
preorder(create_BST(arr,0,var))
```

```python
### BST 3
def inorder_successor(root, x):

    if root==None:
        return None

    if root.elem<=x.elem:
        return inorder_successor(root.right,x)

    if root.elem>x.elem:
        left= inorder_successor(root.left,x)

        if left==None:
            return root
        else:
            return left

#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = n3
print(f'Inorder predecessor of node {x.elem}: {inorder_successor(root, x).elem}') #Inorder predecessor of node 20: 22
```
