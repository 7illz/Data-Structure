```python
class Node:
  def __init__(self,elem,next = None):
    self.elem, self.next = elem,next
```

```python
def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

```python
class Stack:
    def __init__(self):
        self.top=None


    def push(self,val):
        new=Node(val,self.top)
        self.top=new

    def pop(self):
        if self.top==None:
            return None

        elem=self.top.elem
        self.top=self.top.next
        return elem

    def peek(self):
        if self.top==None:
            return None
        return self.top.elem

    def is_empty(self):
        if self.top==None:
            return True
        return False
def print_stack(st):
  if st.is_empty():
    return
  p = st.pop()
  print('|',p,end=' ')
  if p<10:
    print(' |')
  else:
    print('|')
  #print('------')
  print_stack(st)
  st.push(p)
```

```python
st=Stack()
st.push(23)
st.push(9)
st.push(17)
st.push(4)
print(st.peek())
```

```python
def modify_stack(st):
    new=Stack()
    while not st.is_empty():
        if st.top.next==None:
            store=st.pop()
            break
        var=st.pop()
        new.push(var)

    while not new.is_empty():
        temp=new.pop()
        st.push(temp)
    st.push(store)
    return st

var=modify_stack(st)
print_stack(var)


```

```python
def RotateRight(arr):
    temp=arr[len(arr)-1]
    for i in range(len(arr)-1,-1,-1):
        arr[i]=arr[i-1]
    arr[0]=temp
    return arr


arr=np.array([1,2,3,4])
RotateRight(arr)
```
