```python
def remove_compartment(head,n):
    def count_nodes(head):
        cur = head
        count = 0
        while cur!=None:
                count += 1
                cur = cur.next
        return count
    total=count_nodes(head)
    def reverse(head):
        prev = None
        cur = head
        while cur!=None:
            nxt = cur.next
            cur.next = prev
            prev = cur
            cur = nxt
        return prev
    def node_at(head, idx):
        cur=head
        count=0
        while cur!=None:
            if idx==count:
                return cur
            else:
                count+=1
                cur=cur.next
    def index(head,n):
            cur=head
            num=n
            index=0
            while cur!=None:
                if cur.elem!=num:
                    index+=1
                    cur=cur.next
                else:
                    return index
    def remove(nhead,ind):
        if ind >0 and ind < total:
                n1= node_at(nhead,ind-1)
                n2=node_at(nhead,ind+1)
                n1.next=n2
        elif ind==total-1:
                n1=node_at(nhead,ind-1)
                n1.next=None
        return nhead
    temp=head
    count=0
    while temp!=None:
        if temp.elem!=n:
            count+=1
            temp=temp.next
        else:
            break
    if count==total:
        return head
    new_head=reverse(head)
    ind=index(new_head,n)
    call=remove(new_head,ind)
    final=reverse(call)
    return final

print('==============Test Case 1=============')
head = createList(np.array([14,10,15,10,41,10,72]))
number = 10
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head,number)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head) #This should print 14-->10-->15-->10-->41-->72
print()

print('==============Test Case 2=============')
head = createList(np.array([10,15,33,41,14,72]))
number = 10
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head,number)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head) #This should print 15-->33-->41-->14-->72
print()

print('==============Test Case 3=============')
head = createList(np.array([10,15,33,41,14,72]))
number = 56
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head,number)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head) #This should print 10-->15-->33-->41-->14-->72
print()
```
