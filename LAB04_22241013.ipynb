```python
class Patient:
    def __init__(self,id=None,name=None,age=None,bloodgroup=None,next=None,prev=None):
        self.id=id
        self.name=name
        self.age=age
        self.blood_group=bloodgroup
        self.next=next
        self.prev=prev
```

```python
class WRM:
    def __init__(self):
        self.dh=Patient()
        self.dh.next=self.dh
        self.dh.prev=self.dh

    def countNode(self):
        temp=self.dh.next
        count=0
        while temp!=self.dh:
            count+=1
            temp=temp.next
        return count

    def RegisterPatient(self,id, name, age, bloodgroup):

          prev_patient = self.dh.prev
          add_patient = Patient(id, name, age, bloodgroup, self.dh, prev_patient)
          self.dh.prev = add_patient
          prev_patient.next = add_patient
          print()
          print(add_patient.name, "has been registered successfully")

    def ServePatient(self):

        if self.dh.next==self.dh and self.dh.prev==self.dh:
            print('No Patient available to serve')

        else:
                remove_node=self.dh.next
                next_node=remove_node.next
                self.dh.next=next_node
                next_node.prev=self.dh
                print(remove_node.name ,'is being served')
                remove_node=None


    def CancelAll(self):

        if self.dh.next==self.dh :
            return 'Queue is empty from the start'

        else:
            self.dh.next=self.dh
            self.dh.prev=self.dh
            return 'All appointments have been cancelled'

    def CanDoctorGoHome(self):

        if self.dh.next==self.dh and self.dh.prev==self.dh:
            return 'Yes'
        else:
            return "No"
    def ShowAllPatient(self):

        if self.dh.next==self.dh and self.dh.prev==self.dh:
            print('No patient to show')
        else:
            temp=self.dh.next
            name=''
            while temp!=self.dh:
                name+='\n'+ temp.name
                temp=temp.next
            print('The queue is' , name)

    def ReverseTheLine(self):
        if self.dh.next==self.dh :
            print('Not succes')
        elif self.dh.next.next==self.dh:
            print('You need minimum two patients in the register to reverse')
        else:
            temp=self.dh.next
            prev=self.dh
            while temp!=self.dh:
                next=temp.next
                temp.next=prev
                temp.prev=next
                prev=temp
                temp=next
            self.dh.next=prev
            prev.prev=self.dh
            print()
            print('Reverse successfull')
```

```python
obj=WRM()
print('Welcome to Waiting Room Management System:')
while True:
    print("==Choose an Option to proceed further==")
    print("Enter 1 to Register")
    print("Enter 2 to Serve Patient")
    print("Enter 3 to Cancel all appointments")
    print("Enter 4 to Check whether Doctor can Go Home or not")
    print("Enter 5 to See all apppointments")
    print("Enter 6 to Reverse the patient line")
    print("Enter 7 to Exit from the waiting room management system")
    print()
    input_number=int(input('Enter your number: '))
    print()
    if input_number<0 and input_number>7 :
        print('You typed an invalid number')
    elif input_number==1:
        print()
        id=int(input("Enter your id: "))
        name=str(input("Enter your name: "))
        age=int(input('Enter your age: '))
        blood_group=str(input("Enter your bloodgroup: "))
        obj.RegisterPatient(id,name,age,blood_group)
        print()
    elif input_number==2:
        print()
        obj.ServePatient()
        print()
    elif input_number==3:
        print()
        print(obj.CancelAll())
        print()
    elif input_number==4:
        print()
        print(obj.CanDoctorGoHome())
        print()
    elif input_number==5:
        print()
        obj.ShowAllPatient()
        print()
    elif input_number==6:
        print()
        obj.ReverseTheLine()
        print()
    elif input_number==7:
        print()
        print('You have successfully exited our Waiting Room Management System. Thanks for using our System')
        print()
        break



```
