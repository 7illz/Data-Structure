```python
#### factorial
def factorial(n):
    if n==0 or n==1:           #### why n=0
        return 1
    return n * factorial(n-1)

factorial(5)
```

```python
### exponent
def exp(a,n):
    if n==0:
        return 1
    return a * exp(a, n-1)

exp(2,4)
```

```python
##### fibonacci
def fib(n):
    if n==0:
        return 0     ####### or  if n==0 or n==1: return n
    if n==1:
        return 1
    return fib(n-1) + fib(n-2)
fib(6)
fib(7)

```

```python
### maximum
arr=[10,20,-3,50,2,11]
i=0
end=len(arr)-1

def max_of(arr,i,end):
    if i==end:
        return arr[end]
    return max(arr[i],max_of(arr,i+1,end))
max_of(arr,i,end)
```

```python
### sum of
arr=[10,-7,20,30,50]
def sum_arr(arr,i,end):
    if i==end:
        return arr[end]
    return arr[i] + sum_arr(arr,i+1,end)
end=len(arr)-1
sum_arr(arr,0,end)
```

```python
### target sum
arr=[3,3,0,5,10]
def target_sum(arr,i,sum,target):
    if target==sum:
        return True
    if i==len(arr):
        return False
    return target_sum(arr,i+1,sum+arr[i],target) or target_sum(arr,i+1,sum,target)
target_sum(arr,0,0,18)
```

```python
### group sum
arr=[3,3,0,5,10]
def target_sum(arr,i,grp,target):
    print(grp)
    sum=0
    for x in grp:
        sum+=x
    if target==sum:
            return True
    if i==len(arr):
        return False
    return target_sum(arr,i+1,grp+[arr[i]],target) or target_sum(arr,i+1,grp,target)
target_sum(arr,0,[],5)
```

```python
#QUESTION 24
def nestParen(s):

    if len(s)==0:
        return True

    if s[0]=='(' and s[len(s)-1]==')':
        return nestParen(s[1:len(s)-1])

    if s[0]=='(' and s[len(s)-1]!=')' or s[0]!='(' and s[len(s)-1]==')':
        return False

nestParen("(((x))")
```

```python
import math

class Node:
  def __init__(self, elem, next = None):
    self.elem = elem
    self.next = next

def create_linked_list(arr):
  head = Node(arr[0])
  tail = head
  for i in arr[1:]:
    new_node = Node(i)
    tail.next = new_node
    tail = new_node
  return head
```
