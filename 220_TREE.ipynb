```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def preorder(root):
    if root==None:
        return

    print(root.elem, end = ' ')
    preorder(root.left)
    preorder(root.right)

```

```python
def postorder(root):

    if root==None:
        return

    postorder(root.left)
    postorder(root.right)
    print(root.elem, end = ' ')
    return

```

```python
def get_level(root,elem,level):
    if root==None:
        return -1

    if root.elem==elem:
        return level

    left_level=get_level(root.left,elem,level+1)

    if left_level!=-1:
        return left_level

    return get_level(root.right,elem,level+1)

get_level(root2,'C',0)
```
