```python
#Run this cell
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

```python
def count_nodes(head):
   cur = head
   count = 0
   while cur!=None:
        count += 1
        cur = cur.next
   return count
# print(count_nodes(head))
```

```python
def reverse(head):
    prev=None
    cur=head
    while cur!=None:
        next=cur.next
        cur.next=prev
        prev=cur
        cur=next
    return prev
```

```python
### Floyds algorithm
def middle(head):
    slow=head
    fast=head
    while slow!=None and fast!=None and fast.next!=None :
        slow=slow.next
        fast=fast.next.next
    return slow.elem


ar=np.array([10,20,30,40,50])
arr=np.array([1,2,3,4,5,6])
head=createList(ar)
printLinkedList(head)
print(middle(head))
```
